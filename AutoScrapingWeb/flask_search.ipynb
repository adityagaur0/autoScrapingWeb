```python
from autoscraper import AutoScraper
from flask import Flask, request

amazon_scraper = AutoScraper()
amazon_scraper.load('hackathorn-search')
app = Flask(__name__)

def get_amazon_result(query=None):
    url = 'https://devfolio.co/hackathons'
    if query:
        url += f'?q={query}'
    result = amazon_scraper.get_result_similar(url, group_by_alias=True)
    return _aggregate_result(result)

def _aggregate_result(result):
    final_result = []
    for i in range(len(list(result.values())[0])):
        try:
            final_result.append({alias: result[alias][i] for alias in result})
        except:
            pass
    return final_result

@app.route('/', methods=['GET'])
def search_api():
    query = request.args.get('q')
    if query:
        return dict(result=get_amazon_result(query))
    else:
        return dict(result=get_amazon_result())

@app.route('/filter', methods=['GET'])
def filter_api():
    query = request.args.get('q')
    filter_type = request.args.get('filter_type')
    hackathon_name = request.args.get('hackathon_name')

    if hackathon_name:
        filtered_result = [hackathon for hackathon in get_amazon_result() if hackathon.get('Hackathorn_name') == hackathon_name]
        return dict(result=filtered_result)
    elif filter_type:
        filtered_result = [hackathon for hackathon in get_amazon_result(query) if hackathon.get(filter_type) == query]
        return dict(result=filtered_result)
    else:
        return {"error": "Please provide a valid filter type (status, mode, date) or hackathon name"}

@app.route('/upcoming', methods=['GET'])
def upcoming_api():
    upcoming_hackathons = [hackathon for hackathon in get_amazon_result() if hackathon.get('Status') == 'Upcoming']
    return dict(result=upcoming_hackathons)

@app.route('/by_date', methods=['GET'])
def by_date_api():
    date = request.args.get('date')
    if date:
        hackathons_by_date = [hackathon for hackathon in get_amazon_result() if hackathon.get('Date') == date]
        return dict(result=hackathons_by_date)
    else:
        return {"error": "Please provide a valid date"}

@app.route('/by_status', methods=['GET'])
def by_status_api():
    status = request.args.get('status')
    if status:
        hackathons_by_status = [hackathon for hackathon in get_amazon_result() if hackathon.get('Status') == status]
        return dict(result=hackathons_by_status)
    else:
        return {"error": "Please provide a valid status"}

@app.route('/by_participation', methods=['GET'])
def by_participation_api():
    participation = request.args.get('participation')
    if participation:
        hackathons_by_participation = [hackathon for hackathon in get_amazon_result() if hackathon.get('Participation') == participation]
        return dict(result=hackathons_by_participation)
    else:
        return {"error": "Please provide a valid participation"}

@app.route('/by_mode', methods=['GET'])
def by_mode_api():
    mode = request.args.get('mode')
    if mode:
        hackathons_by_mode = [hackathon for hackathon in get_amazon_result() if hackathon.get('Mode') == mode]
        return dict(result=hackathons_by_mode)
    else:
        return {"error": "Please provide a valid mode"}

if __name__ == '__main__':
    app.run(port=8080, host='0.0.0.0')
```
