```python
from autoscraper import AutoScraper
amazon_url="https://devfolio.co/hackathons"

# We can add one or multiple candidates here.
# You can also put urls here to retrieve urls.

wanted_list=["Diversion 2k24","Offline","Open","Starts 03/02/24","+1000 participating"]

scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)

print(scraper.get_result_similar(amazon_url,grouped=True))
```

```python
scraper.get_result_similar("https://devfolio.co/hackathons",grouped=True)
```

```python
scraper.set_rule_aliases({'rule_06qo': 'Hackathorn_name', 'rule_1o1q': 'Mode','rule_fz3t':'Status','rule_4f2x':'Date','rule_9luk':'Participation'})
scraper.keep_rules(['rule_06qo', 'rule_1o1q','rule_fz3t','rule_4f2x','rule_9luk'])
scraper.save('hackathorn-search')
```

```python
scraper.load('hackathorn-search')
```
